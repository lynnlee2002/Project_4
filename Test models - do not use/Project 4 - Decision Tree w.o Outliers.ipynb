```python
import pandas as pd
```

```python
df = pd.read_csv("/Users/jorgedennis/Downloads/Levels_Fyi_Salary_Data.csv")
df.head()
```

```python
#Drop columns where Race and Education are null
df = df.dropna(subset=['Race', 'Education'])
df.head()
```

```python
df = df[df['Unnamed: 32'].isnull()]
df.info()
```

```python
df = df[df['Unnamed: 31'].isnull()]
df.info()
```

```python
df['Unnamed: 30'].unique()
```

```python
df = df.rename(columns={'Unnamed: 30': 'state'})

df.info()
```

```python
df_clean = df[['title', 'totalyearlycompensation', 'yearsofexperience','gender','Masters_Degree', 'Bachelors_Degree', 'Doctorate_Degree',
       'Highschool', 'Some_College', 'Race_Asian', 'Race_White',
       'Race_Two_Or_More', 'Race_Black', 'Race_Hispanic','state']]
df_clean.head()
```

```python
df_encoded = pd.get_dummies(df_clean['state'])
df_clean = pd.concat([df_clean, df_encoded], axis=1)
df_clean.head()
```

```python
df_encoded = pd.get_dummies(df_clean['gender'])
df_clean = pd.concat([df_clean, df_encoded], axis=1)
df_clean.head()
```

```python
df_encoded = pd.get_dummies(df_clean['title'])
df_clean = pd.concat([df_clean, df_encoded], axis=1)
df_clean.head()
```

```python
df_clean = df_clean.drop(['title', 'gender', 'state'], axis=1)

df_clean.head()
```

```python
Q1 = df_clean['totalyearlycompensation'].quantile(0.25)
Q3 = df_clean['totalyearlycompensation'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df_clean[(df_clean['totalyearlycompensation'] < lower_bound) | (df_clean['totalyearlycompensation'] > upper_bound)]
outliers
```

```python
no_outliers = df_clean.drop(df_clean[(df_clean['totalyearlycompensation'] < lower_bound) | (df_clean['totalyearlycompensation'] > upper_bound)].index)
no_outliers.info()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
y = no_outliers['totalyearlycompensation']
X = no_outliers.drop(columns = 'totalyearlycompensation')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model = DecisionTreeRegressor()
model.fit(X_train, y_train)


y_pred = model.predict(X_test)
```

```python
# Calculate mean squared error
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)

# Calculate R-squared score
r2 = r2_score(y_test, y_pred)
print("R-squared Score:", r2)
```
