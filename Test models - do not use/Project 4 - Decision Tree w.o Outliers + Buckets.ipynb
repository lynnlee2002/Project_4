```python
import pandas as pd
```

```python
df = pd.read_csv("/Users/jorgedennis/Downloads/Levels_Fyi_Salary_Data.csv")
df.head()
```

```python
#Drop columns where Race and Education are null
df = df.dropna(subset=['Race', 'Education'])
df.head()
```

```python
df = df[df['Unnamed: 32'].isnull()]
df.info()
```

```python
df = df[df['Unnamed: 31'].isnull()]
df.info()
```

```python
df['Unnamed: 30'].unique()
```

```python
df = df.rename(columns={'Unnamed: 30': 'state'})

df.info()
```

```python
df_clean = df[['title', 'totalyearlycompensation', 'yearsofexperience','gender','Masters_Degree', 'Bachelors_Degree', 'Doctorate_Degree',
       'Highschool', 'Some_College', 'Race_Asian', 'Race_White',
       'Race_Two_Or_More', 'Race_Black', 'Race_Hispanic','state']]
df_clean.head()
```

```python
df_clean.head()
```

```python
df_encoded = pd.get_dummies(df_clean['state'])
df_clean = pd.concat([df_clean, df_encoded], axis=1)
df_clean.head()
```

```python
df_encoded = pd.get_dummies(df_clean['gender'])
df_clean = pd.concat([df_clean, df_encoded], axis=1)
df_clean.head()
```

```python
df_encoded = pd.get_dummies(df_clean['title'])
df_clean = pd.concat([df_clean, df_encoded], axis=1)
df_clean.head()
```

```python
df_clean = df_clean.drop(['title', 'gender', 'state'], axis=1)

df_clean.head()
```

```python
Q1 = df_clean['totalyearlycompensation'].quantile(0.25)
Q3 = df_clean['totalyearlycompensation'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df_clean[(df_clean['totalyearlycompensation'] < lower_bound) | (df_clean['totalyearlycompensation'] > upper_bound)]
outliers
```

```python
no_outliers = df_clean.drop(df_clean[(df_clean['totalyearlycompensation'] < lower_bound) | (df_clean['totalyearlycompensation'] > upper_bound)].index)
no_outliers.info()
```

```python
min_value = no_outliers['totalyearlycompensation'].min()
max_value = no_outliers['totalyearlycompensation'].max()

print("Minimum value:", min_value)
print("Maximum value:", max_value)
```

```python
import pandas as pd

# Create a new DataFrame from the 'no_outliers' DataFrame
new_df = no_outliers.drop(columns=['totalyearlycompensation'])

# Define the bucket intervals and labels
start_value = 10000
increment = 20000
end_value = 450000

bucket_intervals = list(range(start_value, end_value + increment, increment))
bucket_labels = [f'${bucket_intervals[i]//1000}k-${(bucket_intervals[i+1])//1000}k' for i in range(len(bucket_intervals)-1)]

# Add the bucketed 'totalyearlycompensation' column as 'comp_bins'
new_df['comp_bins'] = pd.cut(no_outliers['totalyearlycompensation'], bins=bucket_intervals, labels=bucket_labels, right=False)

# Print the updated DataFrame
print(new_df.head())
```

```python
new_df.info()

```

```python
new_df['comp_bins'].unique()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
X = new_df.drop(columns=['comp_bins'])  # Independent variables
y = new_df['comp_bins']  # Dependent variable
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
# Initialize the Decision Tree model
model = DecisionTreeClassifier()

# Fit the model on the training data
model.fit(X_train, y_train)



y_pred = model.predict(X_test)
```

```python
# Calculate accuracy score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Generate classification report
report = classification_report(y_test, y_pred)
print("Classification Report:\n", report)
```
